## IMDB Dataset.csv (66.2 MB)

```python
import pandas as pd

# Cargar el dataset completo
df_imdb = pd.read_csv('../data/IMDB Dataset.csv')

# Verificar la estructura del dataset
print(df_imdb.info())
print(df_imdb.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='sentiment', data=df_imdb)
plt.title("Distribuci√≥n de Sentimientos (IMDB)")
plt.show()
```

```python
df_imdb['review_length'] = df_imdb['review'].apply(len)
sns.histplot(df_imdb['review_length'], kde=True)
plt.title("Distribuci√≥n de Longitud de Rese√±as (IMDB)")
plt.show()
```
